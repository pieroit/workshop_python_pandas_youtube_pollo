```python
chicken_filter = df['Item'] == 'Chickens'
chicken_filter.value_counts()
```

```python
chickens = df[ chicken_filter ]
chickens['Item'].unique()
```

```python
chickens['Unit'].value_counts()
```

```python
chickens['Value Extended'] = df.loc[chickens['Value']] * 1000
```

```python
chickens[ ['Element', 'Value', 'Unit', 'Value Extended'] ]
```

```python
chickens.to_csv('chickens.csv')
```
